```python
from datetime import date

def diff_dates(date1, date2):
    return abs(date2-date1).days
```

```python
difference=[]
for i in df2.values:
  m1=i[0]
  y1=i[1]
  m2=i[2]
  y2=i[3]
  d1 = date(y1,m1,1)
  d2 = date(y2,m2,1)
  result1 = diff_dates(d2, d1)
  difference.append(result1)
```

```python
df2['processing_days']=difference
```
